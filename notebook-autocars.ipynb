```python
#As all values are showing as False but we do have empty values, it is because our empty values have "?" but we need them to be in NaN form.
#Replacing empty values that contains '?' with NaN so that we can apply data cleaning
df.replace('?',np.nan,inplace=True )
```

```python
#checking 'NaN' values are replaced or not
df.head(10)
```

```python
#Data Cleaning
#Cleaning the data by dropping the whole row if any NaN value comes up in any column.
df.dropna(inplace=True)
#checking
df.head(10)
```

```python
#Converting num-of-doors into integers 
df.replace('four',4,inplace=True)
df.replace('two',2,inplace=True)
```

```python
#printing how many type of strings are occuring in a column
df['num_of_cylinders']. value_counts()
```

```python
#replacing strings with integers in num_of_cylinders
#We are doing this type-casting so that our results should be accurate as we cant calculate stastical properties with strings.
df.replace('five',5,inplace=True)
df.replace('six',6,inplace=True)
df.replace('three',3,inplace=True)
df.replace('eight',8,inplace=True)
```

```python
df.dtypes
```

```python
#converting type of normalized_losses into integers
df['normalized_losses']=df['normalized_losses'].astype('int')
df.dtypes
```

```python
#converting price object type into integer
df['price']=df['price'].astype('int')
df.dtypes
```

```python
#binning of price
bins=np.linspace(min(df["price"]),max(df["price"]),4)
group_names=["low","Medium","High"]
df["price_binned"]=pd.cut(df["price"],bins,labels=group_names,include_lowest=True)
```

```python
#importing matplot library

import matplotlib.pyplot as plt
from matplotlib import pyplot

```

```python
#Making a histogram of division three type of prices
pyplot.bar(group_names, df["price_binned"].value_counts())

# set x/y labels and plot title
plt.pyplot.xlabel("price")
plt.pyplot.ylabel("count")
plt.pyplot.title("priced_binned")
#We can see that low priced are most in count
#medium are 30
#high are 10 only
```
